## Create a Checkpoint

## 'Date'

```python
raw_reason_modified['Date']
```

```python
type(raw_reason_modified['Date'][0])
```

```python
raw_reason_modified['Date'] = pd.to_datetime(raw_reason_modified['Date'], format = '%d/%m/%Y')
```

```python
type(raw_reason_modified['Date'])
```

```python
raw_reason_modified.info()
```

## Extract the Month Value

```python
raw_reason_modified['Date'][0]
```

```python
raw_reason_modified['Date'][0].month
```

```python
list_months = []
list_months
```

```python
for i in range(raw_reason_modified.shape[0]):
    list_months.append(raw_reason_modified['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
raw_reason_modified['Month Value'] = list_months
```

```python
raw_reason_modified
```

## Extract the Day of the Week

```python
raw_reason_modified['Date'][699].weekday()
```

```python
raw_reason_modified['Date'][699]
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
raw_reason_modified['Day of the Week'] = raw_reason_modified['Date'].apply(date_to_weekday)
```

```python
raw_reason_modified.columns.values
```

```python
columns_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Month Value',
       'Day of the Week','Transportation Expense', 'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
raw_reason_modified = raw_reason_modified[columns_reordered]
```

```python
raw_reason_modified
```

```python
raw_reason_modified = raw_reason_modified.copy()
raw_reason_modified
```
