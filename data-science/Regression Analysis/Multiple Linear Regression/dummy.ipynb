```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import sklearn
import scipy

sns.set()
```

```python
data = pd.read_csv('1.03.+Dummies.csv')
data
```

```python
# dummy = imitation of categories with numbers

new_data = data.copy()

new_data['Attendance'] = new_data['Attendance'].map({'Yes':1,'No':0})
new_data
```

```python
new_data.describe()
```

## REGRESSION

```python
y = new_data['GPA']
x1 = new_data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

GPA = 0.6439 + 0.0014 * (SAT) + 0.2226 (Attendance / dummy) </br>

```python
plt.scatter(new_data['SAT'],y)
yhat_no = 0.6439 + 0.0014 * new_data['SAT']
yhat_yes = 0.8665 + 0.0014 * new_data['SAT']
 
fig = plt.plot(new_data['SAT'],yhat_no,lw=2,c='green')
fig = plt.plot(new_data['SAT'],yhat_yes,lw=2,c='red')

plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
